## Agrupando dados

```python
import pandas as pd
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
dados.head()
```

```python
# bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
# selecao = dados['Bairro'].isin(bairros)
# dados = dados[selecao]
```

```python
dados['Bairro'].drop_duplicates()
```

```python
grupo_bairro = dados.groupby('Bairro')
```

```python
grupo_bairro['Valor m2'].aggregate(['mean', 'max'])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20, 10))
```

```python
fig = grupo_bairro['Valor m2'].std().plot.bar(color = 'blue')
fig.set_ylabel('DP do Valor do Aluguel por m2')
fig.set_title('DP do Valor Médio por m2 por Bairro', {'fontsize': 22})
fig
```
